```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import matplotlib.cm as cm
from matplotlib import colormaps
from matplotlib.colors import Normalize
```

```python
all_sheets = pd.read_excel('Baza_Przemytu.xlsx', sheet_name=None)
```

```python
for year in range(2013,2023):
    all_sheets[str(year)].drop(['Uwagi','Informacje dodatkowe', 'Wniosek do UC',], axis=1, inplace=True)
```

Ilość przestępstw w poszczególnych latach

```python
records_count = {}

for sheet_name, df in all_sheets.items():
    records_count[sheet_name] = len(df)


plt.figure(figsize=(10, 6))
plt.bar(records_count.keys(), records_count.values(), color='skyblue')
plt.xlabel('Rok zdarzenia')
plt.ylabel('Liczba przemytów')
plt.title('Ilość przestępstw w poszczególnych latach')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
crime_type = {}

for sheet_name, df in all_sheets.items():
    crime_type[sheet_name] = df['Towar'].value_counts().head(5)
    
    

```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(15, 10))

for i, (year, crime_data) in enumerate(crime_type.items()):
    
    ax = axes[i // 4, i % 4]
    categories = crime_data.index
    colors = plt.cm.get_cmap('viridis', len(categories))
    
    
    crime_data.plot(kind='bar', ax=ax, title=f'Rok - {year}', color=colors(np.arange(len(categories))))

    ax.set_xlabel('Towar')
    ax.set_ylabel('Liczba')
fig.suptitle('5 najpopularniejszych przestępstw w roku', fontsize=16)
plt.tight_layout()
plt.show()
```

Przemyt papierosów na przestrzenii lat

```python
cigarete_per_year = {}

for sheet_name, df in all_sheets.items():
    cigarete_per_year[sheet_name] = df['Towar'].value_counts()['Papierosy']
    
    
plt.figure(figsize=(10, 6))
plt.bar(cigarete_per_year.keys(), cigarete_per_year.values(), color='skyblue')
plt.xlabel('Rok zdarzenia')
plt.ylabel('Liczba przemytów')
plt.title('Ilość przemytu papierosów na przestrzeni lat')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
    
```
